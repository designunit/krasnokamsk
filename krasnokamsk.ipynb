```python
import pandas as pd
import json
```

```python
cols = [
    'М', 'Ж',
    'Взрослые', 'Дети', 'Дошкольники', 'Молодежь', 'Пенсионеры',
    'Едят/пьют',  'Играют', 'Идут', 'Сидят', 
    'С коляской', 'С собакой',
    'Смартфон', 'Велосипед', 'Спорт', 'Стоят', 'Школьники'
]
def fill_df(df):
    for key in cols:
        if key not in df:
            df[key] = 0
    return df
```

```python
def create_df(filename):
    gjs = json.load(open(filename))

    df = pd.DataFrame([f['properties'] for f in gjs['features']])
    df.drop(['dateCreated', 'lastModified', 'Комментарий'], axis=1, inplace=True)
    df['Количество'].fillna(value=1, inplace=True)
    df = df.fillna(value=False)
    df = df.astype('int64')
    return fill_df(df)
```

```python
df1 = create_df('1309 Набережная.geojson')
df1['location'] = 1

df2 = create_df('1309 Спортсквер.geojson')
df2['location'] = 2

df3 = create_df('1309 Планета Звезд.geojson')
df3['location'] = 3

df = pd.concat([df1, df2, df3], sort=True)
```
